```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
from fuzzywuzzy import fuzz
```

```python
train_data = pd.read_csv('datasets\kaggle_startups_train_28062024.csv')
```

```python
train_data_prep = pd.DataFrame(train_data['city'].unique(),columns = ['city'])
train_data_prep
```

```python
import pandas as pd
from rapidfuzz import fuzz, process


# Пороговое значение для определения дублей
threshold = 80

# Поиск дублей
duplicates = []
for i in range(len(train_data_prep)):
    for j in range(i + 1, len(train_data_prep)):
        similarity = fuzz.ratio(train_data_prep['city'][i], train_data_prep['city'][j])
        if similarity > threshold:
            duplicates.append((train_data_prep['city'][i], train_data_prep['city'][j]))

print("Найденные дубли:")
for dup in duplicates:
    print(dup)
```
