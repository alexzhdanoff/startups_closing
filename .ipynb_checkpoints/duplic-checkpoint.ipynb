```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
from fuzzywuzzy import fuzz
```

```python
train_data = pd.read_csv('datasets\kaggle_startups_train_28062024.csv')
train_data.head(5)
```

```python
train_data['full_address'] = train_data[['region', 'city']].astype(str).agg('|'.join, axis=1)

full_address = pd.DataFrame(train_data['full_address'].value_counts().reset_index())
full_address
```

```python
import pandas as pd
from rapidfuzz import fuzz, process


# Пороговое значение для определения дублей
threshold = 90

# Поиск дублей
duplicates = []
for i in range(len(full_address)):
    for j in range(i + 1, len(full_address)):
        similarity = fuzz.ratio(full_address['full_address'][i], full_address['full_address'][j])
        if similarity > threshold:
            duplicates.append((full_address['full_address'][i], full_address['full_address'][j]))

print("Найденные дубли:")
for dup in duplicates:
    print(dup)
```

```python
duplicates_df = pd.DataFrame(duplicates, columns=['Address 1', 'Address 2'])

duplicates_df = pd.merge(duplicates_df, full_address, how='left', left_on='Address 1', right_on="full_address")
duplicates_df = pd.merge(duplicates_df, full_address, how='left', left_on='Address 2', right_on="full_address")
duplicates_df = duplicates_df[['Address 1', 'Address 2', 'count_x', 'count_y']].sort_values(by='count_x', ascending=False)

duplicates_df['crit'] = duplicates_df['count_x'] / duplicates_df['count_y'] 

pd.set_option('display.max_rows', None)
duplicates_df
```
